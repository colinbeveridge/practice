# Word Guessing Game Jupyter Notebook


```python
# this cell loads words and selects a random word from the list for the use to guess
import numpy as np

print('You have 7 guesses to get the word correct. An incorrect guess subtracts a guess.')
print('A correct guess reveals where the letters are in the word.')
print('Guesses should be lowercase letters only.')
print('To be given an option to stop the game, enter "STOP".')

def load_words():
    with open('words_alpha.txt') as word_file:
        valid_words = list(word_file.read().split())

    return valid_words

# record words that were chosen
chosenwords = set()

valid_words = load_words()
length = len(valid_words)

# all below will repeat with each play
# record wins and losses
wins = 0
losses = 0

# start first game
play = True

# get random index
while play:
    print('Begin Game')
    select = np.random.randint(length)
    chosenwords.add(select)

    # get word with index
    word = valid_words[select]
    
    # make list word for assigning
    listword = list(word)

    # initialize showresult and number of guesses
    guesses = 7
    showresult = ['-']*len(word)
    print('Current word status: '+''.join([str(letter) for letter in showresult]))
    while guesses > 0:
        # prompt for first guess
        guess = input('Enter Guess:')

        # check not numeric
        alphabetical = guess.isalpha()
        # allows user to end game whenever they desire
        if guess == 'STOP':
            playletter = 'No'
            break
        # if numeric keep prompting until alphabetical
        while not alphabetical:
            print('Guess must be a letter or word. No numbers or special characters.')
            guess = input('Enter Guess: ')

        # once guess format correct, make lowercase to compare
        guess = guess.lower()
        # check for all instances of guess in word
        locations = [i for i in range(len(word)) if word.startswith(guess, i)]

        # see if there are any instances, if no instances then subtract from guesses and reprompt
        if len(locations) == 0:
            print(guess + ' is not in the word. Try again.')
            guesses -= 1
            print('You have '+ str(guesses) + ' guesses left.')
            # logic if you are out of guesses
            if guesses == 0:
                print('You have run out of guesses. You lose.')
                losses += 1
                # get play again value from user
                break

        else:
            # loop through where guess occurs
            for j in locations:
                # assign letters to result based on what is guessed
                if len(guess) == 1:
                    showresult[j] = listword[j]

                else:
                    showresult[j:j+len(guess)] = listword[j:j+len(guess)]
            # show where the word is at
            print('Current word status: '+''.join([str(letter) for letter in showresult]))
            
            # logic to see if the user has won on that guess
            isComplete = showresult == listword
            if isComplete:
                print('You have correctly guessed the word. You win!')
                wins += 1
                break
    # in outside loop after inner loop breaks, ask if play again
    playletter = input('If you would like to stop playing, type "No". Otherwise, hit any key to play again. R: ')
    if playletter == 'No':
        play = False
        
print('Wins: '+str(wins))
print('Losses: '+str(losses))
print('Game Over')
```
